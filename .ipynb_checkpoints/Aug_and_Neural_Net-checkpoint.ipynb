```python
 def cek_skor(aug_byk,data_ori) :
  from sklearn.cross_decomposition import PLSRegression
  from sklearn.model_selection import cross_val_score,cross_validate
  from sklearn.linear_model import LinearRegression

  column_mapping = {old_name: new_name for old_name, new_name in zip(aug_byk.columns, data_ori.columns)}
  aug_byk=aug_byk.rename(columns=column_mapping)

  new_df = (pd.concat([data_ori,data])[pd.concat([data_ori,data])
                              .duplicated(subset=data.columns, keep=False)]
            .drop_duplicates(subset=data.columns))
  aug_byk=aug_byk.drop(new_df.index)

  x_byk=aug_byk.drop(['Komposisi'], axis=1)
  y_byk=aug_byk['Komposisi']
  x_ori=data_ori.iloc[:, 0:-1]
  y_ori=data_ori.iloc[:,-1]

  plsr = PLSRegression(n_components=15)
  plsr.fit(x_byk,y_byk)
  lv=plsr.transform(x_byk)
  lv=pd.DataFrame(lv)
  lv_ori=plsr.transform(x_ori)
  lv_ori=pd.DataFrame(lv_ori)


  from sklearn.linear_model import LinearRegression
  model_pls= LinearRegression()

  # Fit the model to the data
  model_pls.fit(lv, y_byk)

  cv_scores_model_pls= cross_validate(model_pls, lv, y_byk, cv=9,scoring=['neg_mean_squared_error','r2'])  # 10-fold Average Cross-Validation

  print("PLS-OLS Average Cross-Validation Root Mean Score/ RMSE:", np.sqrt(-cv_scores_model_pls['test_neg_mean_squared_error'].mean()))
  print("PLS-OLS Average Cross-Validation R2:", cv_scores_model_pls['test_r2'].mean())

  # Predict on the original dataset
  y_pred_pls = model_pls.predict(lv_ori)

  # Calculate RMSE and R^2
  rmse_pls = np.sqrt(mean_squared_error(y_ori, y_pred_pls))
  r2_pls = r2_score(y_ori, y_pred_pls)

  print("PLS RMSE on original dataset:", rmse_pls)
  print("PLS R^2 on original dataset:", r2_pls)
  print("")
  print("")
```

```python
def aug_mvn2(data,mult,mean_range,sdv_range):
  import tensorflow as tf
  import random
  from functools import partial
  # Assuming X is your NIRS dataset loaded as a numpy array
  X_aug=data.drop(['Komposisi'], axis=1).values
  y = np.array(data["Komposisi"])

  # Convert your NIRS data and labels into a TensorFlow dataset
  nirs_dataset = tf.data.Dataset.from_tensor_slices((X_aug, y))
  nirs_dataset = nirs_dataset.map(lambda spectrum, label: (tf.cast(spectrum, tf.float32), label))

  # Define the augmentation function
  def augment_data(spectrum, label, mean=0.0, stddev=0.1):
      # Cast spectrum to float32
      spectrum = tf.cast(spectrum, tf.float32)
      # Apply augmentation by adding noise with given mean and stddev
      augmented_spectrum = spectrum + tf.random.normal(tf.shape(spectrum), mean=mean, stddev=stddev)
      return augmented_spectrum, label

  # Map the augmentation function to the dataset
  augmented_dataset1 = nirs_dataset.map(lambda spectrum, label: augment_data(spectrum, label, mean=0, stddev=0.1))
  means = [random.uniform(-mean_range, mean_range) for _ in range(mult)]
  # stddevs=[0.0005* i for i in range(20)]
  stddevs=[random.uniform(-sdv_range,sdv_range) for _ in range(mult)]


  # Number of iterations for augmentation
  num_iterations = mult

  # # Initialize the accumulating dataset with the original dataset
  # accumulating_dataset = nirs_dataset

  # Initialize an empty accumulating dataset with the same structure as nirs_dataset
  output_types = nirs_dataset.element_spec[0].dtype, nirs_dataset.element_spec[1].dtype
  output_shapes = nirs_dataset.element_spec[0].shape, nirs_dataset.element_spec[1].shape
  accumulating_dataset = tf.data.Dataset.from_tensors((tf.zeros(output_shapes[0], output_types[0]), tf.zeros(output_shapes[1], output_types[1]))).take(0)


  for i in range(num_iterations):
      # Map the augmentation function to the dataset
      # Note: Adjust mean and stddev as per your requirement for each iteration
      augmented_dataset = nirs_dataset.map(lambda spectrum, label: augment_data(spectrum, label, mean=means[i], stddev=stddevs[i]))

      # Concatenate the new augmented dataset to the accumulating dataset
      accumulating_dataset = accumulating_dataset.concatenate(augmented_dataset)

  # Optional: Shuffle and batch the final accumulated dataset
  batch_size = 8  # Adjust batch size as needed
  final_dataset = accumulating_dataset.shuffle(buffer_size=1000).batch(batch_size)

  features = np.array([]).reshape(0, 1501) # Adjust 1502 if your feature size is different
  labels = np.array([])

  # Iterate through the dataset and collect all data
  for batch_features, batch_labels in final_dataset:
      # Convert tensors to numpy arrays
      batch_features_np = batch_features.numpy()
      batch_labels_np = batch_labels.numpy()

      # Concatenate the batch data to the existing arrays
      features = np.concatenate((features, batch_features_np), axis=0)
      labels = np.concatenate((labels, batch_labels_np), axis=0)

  # Create a DataFrame from the collected data
  df_features = pd.DataFrame(features)
  df_labels = pd.Series(labels)

  # If you want a single DataFrame with both features and labels
  df = pd.concat([df_features, df_labels.rename("Label")], axis=1)
  return df
```

# Augmentasi jenis 2

```python
import pandas as pd
import numpy as np

def aug_jenis2(data, multi):
    import random
    def offsets_fun(data, offset_range):
        data_copy = data.copy()
        x = data_copy.iloc[:, :-1]
        y = data_copy.iloc[:, -1]
        offsets = np.random.uniform(-offset_range*0.01, offset_range*0.01, size=x.shape[0]) * x.std().mean()*0.001
        df_offset = x.add(offsets, axis=0)
        df_offset = pd.concat([df_offset, y], axis=1)
        return df_offset

    def offsets_sloped_fun(data, grad):
        data_copy = data.copy()
        x = data_copy.iloc[:, :-1]
        y = data_copy.iloc[:, -1]
        slope_range = (-grad, grad)
        slopes = np.random.uniform(slope_range[0]*0.0001, slope_range[1]*0.0001, size=x.shape[0])
        for index, slope in enumerate(slopes):
            x.iloc[index] *= np.linspace(slope_range[0], slope, x.shape[1])
        data_sloped = pd.concat([x, y], axis=1)
        return data_sloped

    def mult_nirs(data, multiplication_range):
        data_copy = data.copy()
        data_to_multiply = data_copy.iloc[:, :-1]
        multiplications = np.random.uniform(1 - multiplication_range*0.01, 1 + multiplication_range*0.01, size=data_to_multiply.shape[0])
        data_mult = data_to_multiply.mul(multiplications, axis=0)
        data_mult_with_last_col = pd.concat([data_mult, data.iloc[:, -1]], axis=1)
        return data_mult_with_last_col

    offset_range = 0.10 * data.std().mean()
    multiplication_range = 1+0.05 * data.std().mean()
    list_offset_range = [offset_range] * multi
    list_grad = [random.uniform(-0.05, 0.05) for _ in range(multi)]
    list_multiplication_range = [multiplication_range] * multi

    result = pd.DataFrame()
    for offset_range, grad, multiplication_range in zip(list_offset_range, list_grad, list_multiplication_range):
        offsets = offsets_fun(data, offset_range)
        sloped = offsets_sloped_fun(data, grad)
        nirs_mult = mult_nirs(data, multiplication_range)
        result = pd.concat([result, offsets, sloped, nirs_mult], ignore_index=True)
    result = result.loc[~((result >= -0.1) & (result <= 0.1)).all(axis=1)]
        # Calculate the absolute sum of each row
    result['abs_sum'] = result.abs().sum(axis=1)

    # Sort the dataframe based on the absolute sum
    result = result.sort_values(by='abs_sum')
    result=result[result['abs_sum'] >= 20]
    result=result.drop('abs_sum',axis=1)
    return result
```

```python
def complete_aug_paper(data,noise,scatter,multiplication):
  #data : Pandas dataframe dengan spectrum dan komposisi
  #noise : list, berisi noise noise yang diinginkan
    #  1 : Normal/Gaussian Distribution Noise
    #  2 : Uniform Distribution Noise
    #  3 : Exponential Distribution Noise

  #scatter : int, berisi metode - metode scatter yang diinginkan
    #  1 : EMSC
    #  2 : SNV
    #  3 : RNV

  from chemotools.scatter import ExtendedMultiplicativeScatterCorrection
  from chemotools.augmentation import NormalNoise
  from chemotools.augmentation import UniformNoise
  from chemotools.augmentation import ExponentialNoise
  from chemotools.augmentation import BaselineShift
  from chemotools.augmentation import IndexShift
  from chemotools.augmentation import SpectrumScale

  import random

  from chemotools.scatter import ExtendedMultiplicativeScatterCorrection
  from chemotools.scatter import StandardNormalVariate
  from chemotools.scatter import RobustNormalVariate

  data=data.reset_index(drop=True)
  spectra=data.iloc[:,:-1].copy()
  label=data.iloc[:,-1].copy()
  spectra.columns = spectra.columns.astype(str)
  list_columns=spectra.columns


  std_list_min=data.std(axis=1).values.min()
  std_list_max=data.std(axis=1).values.max()
  augmented_spectra=pd.DataFrame(columns=spectra.columns)
  for i in range (multiplication):
    stard_dev=np.random.uniform(std_list_min,std_list_max)

    #Augmentation with peak shift noise
    #
    #Augmentation of the spectra by shifting the peak a defined number of indices along the x-axis. This
    #augmentation technique is specially interesting in Raman spectra, where peak shifts between or within
    #instruments can occur as result of differences in the gratings.
    shift = IndexShift(shift=np.random.uniform(-0.1,0.1)*stard_dev)
    shifted_spec = pd.DataFrame(shift.fit_transform(spectra),columns=spectra.columns)
    shifted_spec=pd.concat([shifted_spec,label],axis=1)

    #Augmentation with peak scaling
    #
    #Scales the data by a value drawn from the uniform distribution centered around 1.0 with
    scale = SpectrumScale(scale=np.random.uniform(1-0.1*stard_dev,1+0.1*stard_dev))
    scalled_spec = pd.DataFrame(scale.fit_transform(spectra),columns=spectra.columns)
    scalled_spec=pd.concat([scalled_spec,label],axis=1)

    #Augmentation with baseline shift
    #
    #Adds a baseline to the data. The baseline is drawn from a one-sided uniform distribution between 0 and 0 + scale.
    baseline = BaselineShift(scale=np.random.uniform(0,0.1)*stard_dev)
    baseline_spectra = pd.DataFrame(baseline.fit_transform(spectra),columns=spectra.columns)
    baseline_spectra=pd.concat([baseline_spectra,label],axis=1)

    noise_spectre=pd.DataFrame(columns=spectra.columns)
    if 1 in noise:
      normal_noise = NormalNoise(scale=stard_dev)
      gauss_spectra = pd.DataFrame(normal_noise.fit_transform(spectra),columns=spectra.columns)
      gauss_spectra=pd.concat([gauss_spectra,label],axis=1)
      noise_spectre=pd.concat([noise_spectre,gauss_spectra])
    else:
      pass

    if 2 in noise:
      uniform_noise = UniformNoise(min=-stard_dev*0.1, max=stard_dev*0.1,)
      unif_spectra = pd.DataFrame(uniform_noise.fit_transform(spectra),columns=spectra.columns)
      unif_spectra=pd.concat([unif_spectra,label],axis=1)
      noise_spectre=pd.concat([noise_spectre,unif_spectra])
    else:
      pass

    if 3 in noise:
      exponential_noise = ExponentialNoise(scale=stard_dev)
      exponential_spectra = pd.DataFrame(exponential_noise.fit_transform(spectra),columns=spectra.columns)
      exponential_spectra=pd.concat([exponential_spectra,label],axis=1)
      noise_spectre=pd.concat([noise_spectre,exponential_spectra])
    else:
      pass
    augmented_spectra=pd.concat([augmented_spectra,shifted_spec,scalled_spec,baseline_spectra,noise_spectre])

  if scatter==1:
    emsc = ExtendedMultiplicativeScatterCorrection(use_mean=True)
    final_spectra = pd.DataFrame(emsc.fit_transform(augmented_spectra.iloc[:,:-1]),columns=augmented_spectra.columns[:-1])
    final_spectra=pd.concat([final_spectra,augmented_spectra.iloc[:,-1].reset_index(drop=True)],axis=1,ignore_index=True)
  elif scatter==2:
    snv = StandardNormalVariate()
    final_spectra = pd.DataFrame(snv.fit_transform(augmented_spectra.iloc[:,:-1]),columns=augmented_spectra.columns[:-1])
    final_spectra=pd.concat([final_spectra,augmented_spectra.iloc[:,-1].reset_index(drop=True)],axis=1,ignore_index=True)
  else:
    rnv = RobustNormalVariate()
    final_spectra = pd.DataFrame(rnv.fit_transform(augmented_spectra.iloc[:,:-1]),columns=augmented_spectra.columns[:-1])
    final_spectra=pd.concat([final_spectra,augmented_spectra.iloc[:,-1].reset_index(drop=True)],axis=1,ignore_index=True)

  final_spectra.columns = data.columns


  return final_spectra
```

```python
def complete_aug(data,noise,scatter,multiplication):
  #data : Pandas dataframe dengan spectrum dan komposisi
  #noise : list, berisi noise noise yang diinginkan
    #  1 : Normal/Gaussian Distribution Noise
    #  2 : Uniform Distribution Noise
    #  3 : Exponential Distribution Noise

  #scatter : int, berisi metode - metode scatter yang diinginkan
    #  1 : EMSC
    #  2 : SNV
    #  3 : RNV

  from chemotools.scatter import ExtendedMultiplicativeScatterCorrection
  from chemotools.augmentation import NormalNoise
  from chemotools.augmentation import UniformNoise
  from chemotools.augmentation import ExponentialNoise
  from chemotools.augmentation import BaselineShift
  from chemotools.augmentation import IndexShift
  from chemotools.augmentation import SpectrumScale

  import random

  from chemotools.scatter import ExtendedMultiplicativeScatterCorrection
  from chemotools.scatter import StandardNormalVariate
  from chemotools.scatter import RobustNormalVariate

  data=data.reset_index(drop=True)
  spectra=data.iloc[:,:-1].copy()
  label=data.iloc[:,-1].copy()
  spectra.columns = spectra.columns.astype(str)


  std_list_min=data.std(axis=1).values.min()
  std_list_max=data.std(axis=1).values.max()
  augmented_spectra=pd.DataFrame(columns=spectra.columns)
  for i in range (multiplication):
    stard_dev=np.random.uniform(std_list_min,std_list_max)

    #Augmentation with peak shift noise
    #
    #Augmentation of the spectra by shifting the peak a defined number of indices along the x-axis. This
    #augmentation technique is specially interesting in Raman spectra, where peak shifts between or within
    #instruments can occur as result of differences in the gratings.
    shift = IndexShift(shift=np.random.uniform(-0.01,0.01)*stard_dev)
    shifted_spec = pd.DataFrame(shift.fit_transform(spectra),columns=spectra.columns)
    shifted_spec=pd.concat([shifted_spec,label],axis=1)

    #Augmentation with peak scaling
    #
    #Scales the data by a value drawn from the uniform distribution centered around 1.0 with
    scale = SpectrumScale(scale=np.random.uniform(1-0.001*stard_dev,1+0.001*stard_dev))
    scalled_spec = pd.DataFrame(scale.fit_transform(spectra),columns=spectra.columns)
    scalled_spec=pd.concat([scalled_spec,label],axis=1)

    #Augmentation with baseline shift
    #
    #Adds a baseline to the data. The baseline is drawn from a one-sided uniform distribution between 0 and 0 + scale.
    baseline = BaselineShift(scale=np.random.uniform(0,0.001)*stard_dev)
    baseline_spectra = pd.DataFrame(baseline.fit_transform(spectra),columns=spectra.columns)
    baseline_spectra=pd.concat([baseline_spectra,label],axis=1)

    noise_spectre=pd.DataFrame(columns=spectra.columns)
    if 1 in noise:
      normal_noise = NormalNoise(scale=stard_dev*0.01)
      gauss_spectra = pd.DataFrame(normal_noise.fit_transform(spectra),columns=spectra.columns)
      gauss_spectra=pd.concat([gauss_spectra,label],axis=1)
      noise_spectre=pd.concat([noise_spectre,gauss_spectra])
    else:
      pass

    if 2 in noise:
      uniform_noise = UniformNoise(min=-stard_dev*0.001, max=stard_dev*0.1,)
      unif_spectra = pd.DataFrame(uniform_noise.fit_transform(spectra),columns=spectra.columns)
      unif_spectra=pd.concat([unif_spectra,label],axis=1)
      noise_spectre=pd.concat([noise_spectre,unif_spectra])
    else:
      pass

    if 3 in noise:
      exponential_noise = ExponentialNoise(scale=stard_dev*0.01)
      exponential_spectra = pd.DataFrame(exponential_noise.fit_transform(spectra),columns=spectra.columns)
      exponential_spectra=pd.concat([exponential_spectra,label],axis=1)
      noise_spectre=pd.concat([noise_spectre,exponential_spectra])
    else:
      pass
    augmented_spectra=pd.concat([augmented_spectra,shifted_spec,scalled_spec,baseline_spectra,noise_spectre])

  # if scatter==1:
  #   emsc = ExtendedMultiplicativeScatterCorrection(use_mean=True)
  #   final_spectra = pd.DataFrame(emsc.fit_transform(augmented_spectra.iloc[:,:-1]),columns=augmented_spectra.columns[:-1])
  #   final_spectra=pd.concat([final_spectra,augmented_spectra.iloc[:,-1].reset_index(drop=True)],axis=1,ignore_index=True)
  # elif scatter==2:
  #   snv = StandardNormalVariate()
  #   final_spectra = pd.DataFrame(snv.fit_transform(augmented_spectra.iloc[:,:-1]),columns=augmented_spectra.columns[:-1])
  #   final_spectra=pd.concat([final_spectra,augmented_spectra.iloc[:,-1].reset_index(drop=True)],axis=1,ignore_index=True)
  # else:
  #   rnv = RobustNormalVariate()
  #   final_spectra = pd.DataFrame(rnv.fit_transform(augmented_spectra.iloc[:,:-1]),columns=augmented_spectra.columns[:-1])
  #   final_spectra=pd.concat([final_spectra,augmented_spectra.iloc[:,-1].reset_index(drop=True)],axis=1,ignore_index=True)

  # final_spectra.columns = data.columns

  # return final_spectra
  return augmented_spectra
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
x_augs=all_augmented.iloc[:, :-1].values
y_augs=all_augmented.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(x_augs, y_augs, test_size=0.2, random_state=42)

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_train)
y_augs=y_train

X_val_scaled = scaler.transform(X_test)
y_vals=y_test

X_scaled = X_scaled[..., np.newaxis]
X_val_scaled = X_val_scaled[..., np.newaxis]
```

```python
import tensorflow as tf
import numpy as np
from keras.models import Sequential
from keras.layers import Conv1D, Flatten, Dense, Dropout, GaussianNoise
from keras.optimizers import Adadelta
from keras.losses import Huber
import keras_tuner as kt


# Define the CNN model
def create_cnn(hp):
    model = Sequential()
    hp_activation = hp.Choice('activation', values=['relu', 'selu','linear'])
    hp_kernel_1=hp.Int('conv_1', min_value=15, max_value=100, step=15)
    hp_kernel_2=hp.Int('conv_2', min_value=15, max_value=100, step=15)
    hp_kernel_3=hp.Int('conv_3', min_value=15, max_value=100, step=15)
    hp_kernel_4=hp.Int('conv_4', min_value=15, max_value=100, step=15)
    hp_filter_1=hp.Int('conv_1', min_value=15, max_value=100, step=15)
    hp_filter_2=hp.Int('conv_1', min_value=15, max_value=100, step=15)
    hp_filter_3=hp.Int('conv_1', min_value=15, max_value=100, step=15)
    hp_filter_4=hp.Int('conv_1', min_value=15, max_value=100, step=15)
    hp_layer_1 = hp.Int('layer_1', min_value=64, max_value=100, step=64)
    hp_learning_rate = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])


    model.add(GaussianNoise(0.01, input_shape=(1501, 1)))
    model.add(Conv1D(filters=hp_filter_1, kernel_size=hp_kernel_1, activation=hp_activation))
    model.add(Conv1D(filters=hp_filter_2, kernel_size=hp_kernel_2, activation=hp_activation))
    model.add(Conv1D(filters=hp_filter_3, kernel_size=hp_kernel_3, activation=hp_activation))
    model.add(Conv1D(filters=hp_filter_4, kernel_size=hp_kernel_4, activation=hp_activation))
    model.add(Flatten())
    model.add(Dropout(0.0005))
    model.add(Dense(units=hp_layer_1, activation='selu'))
    model.add(Dense(1, activation='linear'))

    model.compile(optimizer=tf.keras.optimizers.AdamW(learning_rate=hp_learning_rate), loss=Huber(), metrics=['mae'])
    return model
```

```python
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('mae')<0.00005):
      self.model.stop_training = True

callbacks  = myCallback()
stop_early = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=15)
```

```python
tuner = kt.Hyperband(create_cnn,
                     objective='loss',
                     max_epochs=100,
                     factor=3,
                     directory='dir',
                     project_name='x')
tuner.search(X_scaled, y_augs, epochs=50, validation_data=(X_val_scaled, y_vals),batch_size=25, callbacks=[stop_early,callbacks])
```

```python
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
```

```python
model = tuner.hypermodel.build(best_hps)
history = model.fit(X_scaled, y_augs, epochs=200, batch_size=20, validation_data=(X_val_scaled, y_vals))
```
