```python
from Bayesian_Neural_Network import bayesian_neural_network

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

np.random.seed(10120024)
```

# Data Preparation

```python
data = pd.read_csv('processed_data.csv')
undersampled_data = pd.read_csv('undersampled_data.csv')
non_undersampled_data = pd.read_csv('non_undersampled_data.csv')

feature_columns = data.columns[:-1]
target_columns = data.columns[-1]
```

```python
test_undersampled_data = pd.concat([
    undersampled_data.loc[undersampled_data['Class'] == 1, :].head(10),
    undersampled_data.loc[undersampled_data['Class'] == -1, :].head(10)
])

train_undersampled_data = undersampled_data.drop(index=test_undersampled_data.index.values)
```

# Bayesian Neural Network Development

## Trains The Model Using The Undersampled Data

```python
np.random.seed(10120024)

input_layer = [len(feature_columns)]
hidden_layers = [48, 32, 24, 16]
output_layer = [1]

total_epochs = 10
initial_lr = 1e-3
lr_decay_rate = 0.98
batch_size = 50
```

```python
bnn = bayesian_neural_network(input_layer, 
                                hidden_layers, 
                                output_layer, 
                                undersampled_data[feature_columns].values, 
                                undersampled_data[target_columns].values, 
                                batch_size=batch_size,
                                initial_lr=initial_lr, 
                                lr_decay_rate=lr_decay_rate,
                                total_epochs=total_epochs)
```

```python
before_performance = pd.DataFrame({
                'Predictions':[bnn.bnn_fp.forward_propagation(feature_data_i, bnn.m, bnn.v, bnn.model_structure)[-2][-1][0, 0] for feature_data_i in bnn.feature_data],
                'Actual':bnn.target_data.reshape(-1, 1).T[0]
            })

before_performance['Type'] = 'Before'

sns.boxplot(data=before_performance, x='Actual', y='Predictions')
```

```python
bnn.train_model()
```

```python
bnn.visualize_model_performance()
```

```python
after_performance = pd.DataFrame({
                'Predictions':[bnn.bnn_fp.forward_propagation(feature_data_i, bnn.m, bnn.v, bnn.model_structure)[-2][-1][0, 0] for feature_data_i in bnn.feature_data],
                'Actual':bnn.target_data.reshape(-1, 1).T[0]
            })

after_performance['Type'] = 'After'
after_performance = pd.concat((after_performance, before_performance))

sns.boxplot(data=after_performance, x='Actual', y='Predictions', hue='Type')
```

## Evaluate Model's Performance on Non-Undersampled Data

```python
mean_pred = []
var_pred = []

for feature_data_i in non_undersampled_data[feature_columns].tail(500).values.reshape(-1, 2, 1):
    _, _, _, _, _, _, _, mz, vz = bnn.bnn_fp.forward_propagation(feature_data_i, bnn.m, bnn.v, bnn.model_structure)

    mean_pred.append(mz[-1][0, 0])
    var_pred.append(vz[-1][0, 0])
```

```python
final_performance = pd.DataFrame({
                        'Mean Predictions':mean_pred,
                        'Var Predictions':var_pred
                    })

final_performance['Binary Predictions'] = bnn.bnn_fp._binary_classification_output_activation_function(final_performance['Mean Predictions'].values)
final_performance['Actual'] = 1
```

```python
final_performance
```

```python
final_performance.loc[final_performance['Binary Predictions'] == final_performance['Actual'], :] \
            .sort_values('Var Predictions')
```

```python
final_performance.describe()
```

```python
sns.boxplot(data=final_performance, x='Actual', y='Mean Predictions')
```

```python
tp, fp, fn, tn  = confusion_matrix(final_performance['Binary Predictions'].values, final_performance['Actual'].values, labels=[-1, 1]).ravel()

# calculate the evaluation metrics for the model
accuracy = (tp + tn) / (tp + tn + fp + fn) * 100
precision = tp / (tp + fp) * 100
recall = tp / (tp + fn) * 100

for eval, text in zip([accuracy, precision, recall], ['Accuracy', 'Precision', 'Recall']):
    print(f'{text} : {np.round(eval, 3)}%')
```
