# Model Performance On Singular Label

## Model Performance on Predicting Fraudulent Label

```python
feature_data = np.random.normal(loc=-1, scale=1, size=(2, 100)).T
target_data = np.ones(100).reshape(-1, 1) 
```

```python
input_layer = [2]
hidden_layers = [8, 4, 2]
output_layer = [1]

total_epochs = 100
initial_lr = 1e-2
end_lr = 1e-4
batch_size = 5
```

```python
bnn = bayesian_neural_network(input_layer, 
                                hidden_layers, 
                                output_layer, 
                                feature_data, 
                                target_data, 
                                batch_size=batch_size,
                                initial_lr=initial_lr, 
                                end_lr=end_lr,
                                total_epochs=total_epochs)
```

```python
np.array([bnn.bnn_fp.forward_propagation(feature_data_i, bnn.m, bnn.v, bnn.model_structure)[-2][-1][0, 0] for feature_data_i in bnn.feature_data])
```

```python
bnn.train_model()
```

```python
bnn.visualize_model_performance()
```

```python
np.array([bnn.bnn_fp.forward_propagation(feature_data_i, bnn.m, bnn.v, bnn.model_structure)[-2][-1][0, 0] for feature_data_i in bnn.feature_data])
```

# Model Performance On Binary Label

```python
feature_data_1 = np.abs(np.random.normal(loc=1, scale=0.5, size=(2, 50)).T)
target_data_1 = np.ones(50).reshape(-1, 1) 

feature_data_2 = np.abs(np.random.normal(loc=0, scale=0.5, size=(2, 50)).T)
target_data_2 = np.ones(50).reshape(-1, 1) * -1

feature_data = np.concatenate((feature_data_1, feature_data_2))
target_data = np.concatenate((target_data_1, target_data_2))
```

```python
input_layer = [2]
hidden_layers = [16, 8, 4]
output_layer = [1]

total_epochs = 250
initial_lr = 1e-3
end_lr = 1e-4
batch_size = 10
```

```python
# np.random.seed(10120024)
bnn = bayesian_neural_network(input_layer, 
                                hidden_layers, 
                                output_layer, 
                                feature_data, 
                                target_data, 
                                batch_size=batch_size,
                                initial_lr=initial_lr, 
                                end_lr=end_lr,
                                total_epochs=total_epochs)
```

```python
np.array([bnn.bnn_fp.forward_propagation(feature_data_i, bnn.m, bnn.v, bnn.model_structure)[-2][-1][0, 0] for feature_data_i in bnn.feature_data[(bnn.target_data == 1).T[0, 0]]])
```

```python
np.array([bnn.bnn_fp.forward_propagation(feature_data_i, bnn.m, bnn.v, bnn.model_structure)[-2][-1][0, 0] for feature_data_i in bnn.feature_data[(bnn.target_data == -1).T[0, 0]]])
```

```python
bnn.train_model()
```

```python
bnn.visualize_model_performance()
```

```python
np.array([bnn.bnn_fp.forward_propagation(feature_data_i, bnn.m, bnn.v, bnn.model_structure)[-2][-1][0, 0] for feature_data_i in bnn.feature_data[(bnn.target_data == 1).T[0, 0]]])
```

```python
np.array([bnn.bnn_fp.forward_propagation(feature_data_i, bnn.m, bnn.v, bnn.model_structure)[-2][-1][0, 0] for feature_data_i in bnn.feature_data[(bnn.target_data == -1).T[0, 0]]])
```
